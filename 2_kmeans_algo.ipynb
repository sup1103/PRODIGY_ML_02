IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

LOADING THE DATA

```python
df = pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.drop(['CustomerID', 'Gender', 'Age'] , axis = 'columns' , inplace  = True)
```

```python
df['Annual Income (k$)'].plot(kind = 'box')
```

SCALING THE COLUMNS

```python
scaler = StandardScaler()
```

```python
scaled_array = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_array, columns = df.columns)
```

```python
scaled_df.head()
```

ELBOW POINT

```python
sse = []
k_range = range(1,11)
for i in k_range:
  km = KMeans(n_clusters = i)
  km.fit (df)
  sse.append(km.inertia_)
```

```python
plt.figure(figsize = (10 ,6))
plt.plot(range(1 , 11) , sse)
plt.plot(range(1, 11), sse, linewidth=2, color="blue", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1, 11, 1))
plt.ylabel("sse")
plt.show()
```

PERFORMING K-MEANS CLUSTERING

```python
km = KMeans(n_clusters = 5 , init = 'k-means++')
y = km.fit_predict(scaled_df)
```

```python
scaled_df["label"] = y
```

```python
scaled_df.head()
```

VISUALIZATION OF CLUSTERING

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', hue = "label", palette = ['green', 'orange', 'brown', 'dodgerblue', 'red'], legend = 'full', data = scaled_df, s = 60)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Spending Score (1-100) vs Annual Income (k$)')
plt.show()
```
